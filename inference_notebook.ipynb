```python
# import joblib
import joblib
# import pandas
import pandas as pd
```

```python
# Load pipeline dari file
basket_model = joblib.load('basketInjModelV1.pkl')
basket_model
```

```python
df_inf = pd.read_csv('data_inf.csv')
# rename kolom agar dapat dibaca oleh model, karna pada pembuatan model jg dilakukan rename kolom
df_inf.columns = ['Player ID', 'Posisi', 'Tipe Cedera', 'Tingkat Cedera',
                'Program Rehab', 'Tanggal Cedera', 'Umur', 'Tinggi', 
                'Berat', 'Durasi Rehab (dalam minggu)',
                'Rehabilitasi Skor', 'Sudut Lutut', 
                'Tinggi Lompatan', 'Sudut Fleksibel Angkle',
                'Kecepatan Gerak', 'Kecepatan Reflek']
df_inf
```

```python
# melakukan prediksi model pada data_inf
potensi_kambuh = basket_model.predict(df_inf)
# menambahkan kolom baru untuk menampung hasil prediksi
df_inf['Potensi Kambuh'] = potensi_kambuh
df_inf
```
